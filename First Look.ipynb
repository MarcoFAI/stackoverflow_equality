```python
# import library and dataset

import pandas as pd
import seaborn as sns

df = pd.read_csv('./data/stack-overflow-developer-survey-2021/2021 Stack Overflow Survey Responses.csv', encoding="ISO-8859-1")

df.head()
```

```python
# show shape and columns of dataset

print(df.shape)
print(df.columns)
```

```python
# get correlation of columns

sns.heatmap(df.corr(), annot=True, fmt=".2f");

# only 3 numerical columns, only correlation between Compensation and Yearly Compensation -> useless
```

```python
# calculate percentage of missing values per column

percent_missing = df.isnull().sum() * 100 / len(df)
percent_missing.sort_values(ascending=False)
```
